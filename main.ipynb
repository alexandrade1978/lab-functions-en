## 1. Write a function that returns the greater of two numbers

```python
def greater(a,b):
    # Returns the greater of two numbers, b value in case of equality or None if any input is None
    if a is None or b is None:
        return None
    return a if a > b else b
```

## 2. Now write a function that returns the largest element on a list

```python
def greatest(lst):
    # Returns the greatest number in a list or None if the list is empty
    if not lst: return None
    bigger = lst[0]
    for num in lst:
        if num > bigger:
            bigger = num
    return bigger
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    # Returns the factorial of a number or None if input number is negative
    if n < 0: return None
    if n == 0: return 1
    result = 1
    for i in range(1, n + 1): 
        result *= i
    return result
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ðŸ¤”`

```python
def unique(lst_un):
    # Returns a list of unique elements from the input list or None if the list is empty
    if not lst_un: return None
    un_elements = []
    for element in lst_un: 
        seen = False
        for unique in un_elements:
            if element == unique:
                seen = True
                break
        if not seen:
            un_elements.append(element)
    return un_elements
```

```python
# This will test your function 
test_unique(unique)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
    # Returns the mode of the input list or None if the input list is empty
    if not arr: return None
    frequency = {}
    for num in arr:
        for num_key in frequency:
            if num == num_key:
                frequency[num_key] += 1
                break
        else:
            frequency[num] = 1
    mode = arr[0]
    for num_key in frequency:
        if frequency[num_key] > frequency[mode]:
            mode = num_key
    return mode
    
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def st_dev(list_sd):
    # Returns the standard deviation of a list or None if the list is empty
    if not list_sd: return 0
    sum_lst = 0
    pop_size = 0
    avg_pop = 0
    sum_tot = 0
    std_dev = 0.0
    for val in list_sd:
        sum_lst += val
        pop_size += 1
    avg_pop = sum_lst / pop_size
    for val in list_sd:
        sum_tot += (val - avg_pop) * (val - avg_pop)
    std_dev = (sum_tot / (pop_size - 1))**0.5
    return std_dev
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    # Returns True if the input string is a pangram, False otherwise
    if not string: return False
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    alphabet_up ="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    string_lower = ""
    # Convert all uppercase letters to lowercase
    for char in string:
        convert = False
        count_up = 0
        for let_up in alphabet_up:
            if char == let_up:
                string_lower += alphabet[count_up]
                convert = True
                break
            count_up += 1
        if convert == False:
            string_lower += char
    # Search for each letter in the alphabet
    for let in alphabet:
        found = False
        for char in string_lower:
            if let == char:
                found = True
                break
        if not found:
            return False
    return True
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
    
    if not string: return None
    words = []
    tmp_word = ""
    for char in string:
        if char == ",":
            words.append(tmp_word)
            tmp_word = ""
        else: tmp_word += char
    if tmp_word: words.append(tmp_word)
    ordered_words = sorted(words)
    ord_string = ""
    for word in ordered_words:
        if ord_string == "":
            ord_string += word
        else: ord_string += "," + word
    return ord_string
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    # Returns True if the password meets the criteria, False otherwise
    if not password: return False
    
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False
    
    # Define character sets for manual validation
    special_chars = "!@#$%^&*()}{][-+?_=,<>/.#"
    upper_chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    lower_chars = "abcdefghijklmnopqrstuvwxyz"
    digit_chars = "0123456789"
    
    len_pass = 0
    for char in password:
        len_pass += 1
    
    # If it's too short, no need to check the rest
    if len_pass < 8: return False

    # Validate each character in the password
    for char in password:
        
        found_type = False
        for up_char in upper_chars:
            if char == up_char:
                has_upper = True
                found_type = True
                break
        if not found_type:
            for low_char in lower_chars:
                if char == low_char:
                    has_lower = True
                    found_type = True
                    break
        if not found_type:
            for dig_char in digit_chars:
                if char == dig_char:
                    has_digit = True
                    found_type = True
                    break
        if not found_type:
            for spec_char in special_chars:
                if char == spec_char:
                    has_special = True
                    found_type = True
                    break    

    # Final check: return True only if all criteria are met
    if has_upper and has_lower and has_digit and has_special:
        return True
    return False
    
```

```python
# This will test your function 
test_pass(check_pass)
```
